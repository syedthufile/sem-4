# morphology and image segmentation

```python
import cv2 as cv
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
img = cv.imread("C:/Users/SN131/Desktop/goku22.jpg")
img = cv.threshold(img,220,225,cv.THRESH_BINARY)
img
```

```python
img = cv.imread("C:/Users/SN131/Desktop/goku22.jpg")
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
_,mask = cv.threshold(img,220,225,cv.THRESH_BINARY)
img
```

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv.dilate(mask,kernel,iterations = 1)
titles = ['image','mask','dilate']
image = [img,mask,dilation]
for i in range(len(image)):
    plt.subplot(1,3,i+1),plt.imshow(image[i])
    plt.title(titles[i])
    plt.xticks([])
```

# morphological opening

```python
open = cv.morphologyEx(img,cv.MORPH_OPEN,kernel)
```

```python
cv.imshow("image",img)
cv.imshow("open",open)
cv.waitKey(0)
cv.destroyAllWindows()
```

# morphological close followed by dilation and erosion

```python
close = cv.morphologyEx(img,cv.MORPH_CLOSE,kernel)
```

```python
cv.imshow("image",img)
cv.imshow("close",close)
cv.waitKey(0)
cv.destroyAllWindows()
```

# morphological gradient

```python
gradient = cv.morphologyEx(img,cv.MORPH_GRADIENT,kernel)
```

```python
cv.imshow("image",img)
cv.imshow("gradient",gradient)
cv.waitKey(0)
cv.destroyAllWindows()
```

# morphological openhat

```python
tophat = cv.morphologyEx(img,cv.MORPH_TOPHAT,kernel)
```

```python
cv.imshow("image",img)
cv.imshow("tophat",tophat)
cv.waitKey(0)
cv.destroyAllWindows()
```

# morphological blackhat

```python
blackhat = cv.morphologyEx(img,cv.MORPH_BLACKHAT,kernel)
```

```python
cv.imshow("image",img)
cv.imshow("blackhat",blackhat)
cv.waitKey(0)
cv.destroyAllWindows()
```
